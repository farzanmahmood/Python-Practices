```python
import pandas as pd
dataSet1 = [1,3,5,7,9]
dataSet2 = [2,4,6,8,10]

series1 = pd.Series(dataSet1)
series2 = pd.Series(dataSet2)
series3 = series1 + series2
print("pandas Series1: ")
print(series1, '\n')
print("pandas Series2: ")
print(series2, '\n')
print("pandas Series3: ")
print(series3)
```
